```python
import pandas as pd
# Import csv
df = pd.read_csv('C:/Analytics/data/buzzWords.csv')

df.head()
```

```python
# First, let's build our dataframe
myWordsDf = pd.DataFrame(columns = ['Words', 'Topic', 'Rank', 'Language'])
```

```python
# Ok, so let's order by topic 1, highest to lowest
df.sort_values(by=['Topic 1'], ascending=False, inplace=True)
df.head(25)                                                    
```

```python
# Add features to topic 1
count = 0
while count < 25:
    myWordsDf = myWordsDf.append({'Words': df.iloc[count]['Feature'], 'Topic': 'Topic 1', 'Rank': count + 1, 'Language': '',}, ignore_index=True)
    count += 1
```

```python
# Now, Topic 2
df.sort_values(by=['Topic 2'], ascending=False, inplace=True)
df.head(25) 
```

```python
# Now, add feature for Topic 2
count = 0
while count < 25:
    myWordsDf = myWordsDf.append({'Words': df.iloc[count]['Feature'], 'Topic': 'Topic 2', 'Rank': count + 1, 'Language': '',}, ignore_index=True)
    count += 1
myWordsDf.head(50)
```

```python
# Now, let's sort by topic 3
df.sort_values(by=['Topic 3'], ascending=False, inplace=True)
df.head(25)
```

```python
# Grab the top 10 features
count = 0
while count < 25:
    myWordsDf = myWordsDf.append({'Words': df.iloc[count]['Feature'], 'Topic': 'Topic 3', 'Rank': count + 1, 'Language': '',}, ignore_index=True)
    count += 1
myWordsDf.head(50)
```

```python
# Next, topic 4
df.sort_values(by=['Topic 4'], ascending=False, inplace=True)
df.head(25)
```

```python
count = 0
while count < 25:
    myWordsDf = myWordsDf.append({'Words': df.iloc[count]['Feature'], 'Topic': 'Topic 4', 'Rank': count + 1, 'Language': '',}, ignore_index=True)
    count += 1
myWordsDf.head(50)
```

```python
# Next, topic 5
df.sort_values(by=['Topic 5'], ascending=False, inplace=True)
df.head(25)
```

```python
count = 0
while count < 25:
    myWordsDf = myWordsDf.append({'Words': df.iloc[count]['Feature'], 'Topic': 'Topic 5', 'Rank': count + 1, 'Language': '',}, ignore_index=True)
    count += 1
myWordsDf.head(125)
```

```python
# Let's take out all of the | symbols in our text:
myWordsDf['Words'] = myWordsDf['Words'].replace('|', '')
myWordsDf
```

```python
from langdetect import detect
from langdetect import DetectorFactory

language = []
DetectorFactory.seed = 0
for words in myWordsDf['Words']:
    language.append(detect(words))
    count += 1

#print(language)
myWordsDf['Language'] = language
```
