```python
# Data setup
# First, upload the data set
import pandas as pd

pd.set_option('display.max_rows', 10000)
pd.set_option('display.max_columns', None)

df = pd.read_csv('C:/Analytics/data/campaigns.csv')

# take year of created date, drop every row that wasn't created in 2020
year = df.created_at.str[:4]
df['ld_year'] = year 
df.drop(df[df['ld_year'] != '2020'].index, inplace=True)
df.drop(columns=['ld_year'], inplace=True)

# make dollar conversions
df['current_amount_USD'] = 0
df.loc[df['currencycode'] == 'USD', 'current_amount_USD'] = df.current_amount
df.loc[df['currencycode'] == 'EUR', 'current_amount_USD'] = df.current_amount*0.92  #conversion rates at point in time
df.loc[df['currencycode'] == 'GBP', 'current_amount_USD'] = df.current_amount*0.81
df.loc[df['currencycode'] == 'CAD', 'current_amount_USD'] = df.current_amount*1.4
df.loc[df['currencycode'] == 'AUD', 'current_amount_USD'] = df.current_amount*1.62
df.loc[df['currencycode'] == 'CHF', 'current_amount_USD'] = df.current_amount*0.97
df.loc[df['currencycode'] == 'SEK', 'current_amount_USD'] = df.current_amount*10.03
df.loc[df['currencycode'] == 'NOK', 'current_amount_USD'] = df.current_amount*10.24
df.loc[df['currencycode'] == 'DKK', 'current_amount_USD'] = df.current_amount*6.85

df['goal_USD'] = 0
df.loc[df['currencycode'] == 'USD', 'goal_USD'] = df.goal
df.loc[df['currencycode'] == 'EUR', 'goal_USD'] = df.goal*0.92  #conversion rates at point in time
df.loc[df['currencycode'] == 'GBP', 'goal_USD'] = df.goal*0.81
df.loc[df['currencycode'] == 'CAD', 'goal_USD'] = df.goal*1.4
df.loc[df['currencycode'] == 'AUD', 'goal_USD'] = df.goal*1.62
df.loc[df['currencycode'] == 'CHF', 'goal_USD'] = df.goal*0.97
df.loc[df['currencycode'] == 'SEK', 'goal_USD'] = df.goal*10.03
df.loc[df['currencycode'] == 'NOK', 'goal_USD'] = df.goal*10.24
df.loc[df['currencycode'] == 'DKK', 'goal_USD'] = df.goal*6.85
```

```python
# Make amount per donor column
df['amount_per_donor'] = df['current_amount_USD']/df['donators']
df.loc[df['donators'] == 0, 'amount_per_donor'] = 0
df.loc[df['current_amount_USD']== 0, 'amount_per_donor'] = 0

# Make the percent of goal column
df['percent_of_goal'] = df['current_amount']/df['goal']
```

```python
# adjust columns for skewness and outliers
# Based on this, we will keep the following:
#               percent of goal:                   cube root
#               amount_per_donor:                  log2

import numpy as np

# Change any crazy outliers in percent of goal to be just 1
df.loc[df['percent_of_goal'] > 1, 'percent_of_goal' ] = 1

df['pergoal_sqrt'] = df['percent_of_goal']**(1/2)
df['pergoal_cubert'] = df['percent_of_goal']**(1/3)
df['pergoal_log2'] = np.log2(df['percent_of_goal'] + 1)
df['pergoal_log10'] = np.log10(df['percent_of_goal'] + 1)
df['pergoal_log'] = np.log(df['percent_of_goal'] + 1)

df['amt_donor_sqrt'] = df['amount_per_donor']**(1/2)
df['amt_donor_cubert'] = df['amount_per_donor']**(1/3)
df['amt_donor_log2'] = np.log2(df['amount_per_donor'] + 1)
df['amt_donor_log10'] = np.log10(df['amount_per_donor'] + 1)
df['amt_donor_log'] = np.log(df['amount_per_donor'] + 1)

#df['currUSD_sqrt'] = df['current_amount_USD']**(1/2)
#df['currUSD_cubert'] = df['current_amount_USD']**(1/3)
#df['currUSD_log2'] = np.log2(df['current_amount_USD'] + 1)
#df['currUSD_log10'] = np.log10(df['current_amount_USD'] + 1)
#df['currUSD_log'] = np.log(df['current_amount_USD'] + 1)

#df['donors_sqrt'] = df['donators']**(1/2)
#df['donors_cubert'] = df['donators']**(1/3)
#df['donors_log2'] = np.log2(df['donators'] + 1)
#df['donors_log10'] = np.log10(df['donators'] + 1)
#df['donors_log'] = np.log(df['donators'] + 1)

print(df['percent_of_goal'].skew())
print('\nPercent of Goal:')
print(df['percent_of_goal'].skew())
print('Square:' + str(df['pergoal_sqrt'].skew()))
print('Cube: ' + str(df['pergoal_cubert'].skew()))
print('Log2: ' + str(df['pergoal_log2'].skew()))
print('Log10: ' + str(df['pergoal_log10'].skew()))
print('Ln: ' + str(df['pergoal_log'].skew()))
print('\nAmount per donor:')
print(df['amount_per_donor'].skew())
print('Square:' + str(df['amt_donor_sqrt'].skew()))
print('Cube: ' + str(df['amt_donor_cubert'].skew()))
print('Log2: ' + str(df['amt_donor_log2'].skew()))
print('Log10: ' + str(df['amt_donor_log10'].skew()))
print('Ln: ' + str(df['amt_donor_log'].skew()))

# Delete unneeded columns
df.drop(columns=['pergoal_sqrt', 'pergoal_log2', 'pergoal_log10', 'pergoal_log', 'amt_donor_sqrt', 'amt_donor_cubert', 'amt_donor_log10', 'amt_donor_log'], inplace=True)
```

```python
# Calculate the mu (mean of the entire population) for each of our four factors:
df['pergoal_score'] = (((df['pergoal_cubert']-df['pergoal_cubert'].mean())/df['pergoal_cubert'].std())*10) + 50
df['amt_donor_score'] = (((df['amt_donor_log2']-df['amt_donor_log2'].mean())/df['amt_donor_log2'].std())*10) + 50
df['success_score'] = (df['pergoal_score'] + df['amt_donor_score'])/2

print('Percent of goal score: ')
print('Mean: ' + str(df['pergoal_score'].mean()))
print('Min: ' + str(df['pergoal_score'].min()))
print('Median: ' + str(df['pergoal_score'].median()))
print('Max: ' + str(df['pergoal_score'].max()))
print('\nAmount per donor score: ')
print('Mean: ' + str(df['amt_donor_score'].mean()))
print('Min: ' + str(df['amt_donor_score'].min()))
print('Median: ' + str(df['amt_donor_score'].median()))
print('Max: ' + str(df['amt_donor_score'].max()))
print('\nSuccess score: ')
print('Mean: ' + str(df['success_score'].mean()))
print('Min: ' + str(df['success_score'].min()))
print('Median: ' + str(df['success_score'].median()))
print('Max: ' + str(df['success_score'].max()))
```
